```python
import csv
import numpy as np
import pylab as pl
from biosppy import storage
from biosppy.signals import ecg

root = '/content/drive/My Drive/Colab Notebooks/'
def importTimeSeries(filename):
    datafile = open(root+filename, 'r')
    datareader = csv.reader(datafile)
    out = dict()
    first = True
    for row in datareader:
        if first:
            first = False
            continue
        out[int(row[0])] = np.array([ float(cell) for cell in row[1:]])
    return out

# each one of these is a dict 
# - key:   is the id
# - value: is the actual time series in np.array format (lengths differ between samples)
X_train = importTimeSeries('X_train.csv')
X_test = importTimeSeries('X_test.csv')
y_train = np.loadtxt('y_train.csv',skiprows=1,delimiter=',')
sample = np.loadtxt('sample.csv',skiprows=1,delimiter=',')
n_train = len(X_train)
n_test = len(X_test)
```
